```python
driver = webdriver.Firefox()
driver.implicitly_wait(30)
driver.maximize_window()
```

```python
kommentare = []

class Kommentar:
    def __init__(self, url, datum, text):
        self.url = url
        self.datum = datum
        self.text = text
```

```python
def kommentar_getter():
    #main
    html = driver.page_source
    soup = BeautifulSoup(html, "html.parser")
    blib = soup.find_all(class_=re.compile("col-right.flex-1"))
    for ding in blib:
        try:
            datum = ding.find(class_=re.compile("watson-discussion__comment__info text-xxs")).get_text()
            textatti = ding.find(class_=re.compile("watson-snippet__text mb-2")).get_text()
            crawled_kom = Kommentar(page, datum, textatti)
            kommentare.append(crawled_kom)
        except AttributeError:
            pass
```

```python
site = "https://www.watson.ch/"
driver.get(site)

cookies = pickle.load(open("watson_cookies.pkl", "rb"))

for cookie in cookies:
    driver.add_cookie(cookie)
    
driver.refresh()
```

```python
with open("watson_output.bin", "rb") as data:
    pages = pickle.load(data)
```

```python
for page in pages[895:1160]:
    time.sleep(4)
    driver.get(page)
    try:
        x = driver.find_element_by_class_name("font-semibold.rounded")
        x.click()
        kommentar_getter()
    except NoSuchElementException:
        pass
```

```python
with open('watson', 'w', newline='', encoding="utf8") as csvfile:
    blogwriter_kom = csv.writer(csvfile, delimiter=';', quotechar='"')
                                
    for element in kommentare:
        blogwriter_kom.writerow( [element.url, element.datum, element.text] )
```
