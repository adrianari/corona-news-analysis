```python
import pandas as pd
import numpy as np
import pickle
```

```python
#creating one frame from the different article csv-files
blick1 = pd.read_csv("blick_art1.csv", delimiter = "|", header = None)
blick2 = pd.read_csv("blick_art2.csv", delimiter = "|", header = None)
blick3 = pd.read_csv("blick_art3.csv", delimiter = "|", header = None)
blick4 = pd.read_csv("blick_art4.csv", delimiter = "|", header = None)

frames = [blick1, blick2, blick3, blick4]

blick_main = pd.concat(frames, ignore_index= True)
```

```python
blick_main[4] = blick_main[4].fillna("Agentur")
```

```python
# data cleaning 
blick_main[4] = blick_main[4].replace(to_replace=r'<div class="detail__item__header__item">', value='', regex=True)
blick_main[4] = blick_main[4].replace(to_replace=r'</div>', value='', regex=True)
```

```python
#drop duplicates
blick_main = blick_main.drop_duplicates(keep = "last")
```

```python
#set date
blick_main[2] = pd.to_datetime(blick_main[2], format="%d.%m.%Y")
blick_main[9] = blick_main[2].dt.year
blick_main1 = blick_main[blick_main[9] != 2021]
blick_main1 = blick_main1.reset_index()
```

```python
# extract categories
caty = blick_main1[6]
catyy = caty.str.split("/",expand=True,)
blick_main1[7] = catyy[3]
catyy[8] = catyy[4]
catyy[9] = catyy[8].str.len()
catyy[9] = catyy[9].astype(int)

catyy[10] = np.where(catyy[9] < 20, catyy[4],"")

blick_main1[7] = catyy[3]
blick_main1[8] = catyy[10]

blick_main1 = blick_main1.drop(['index'], axis=1)
blick_main.reset_index(drop = True)
blick_main1
```

```python
blick_main1 = blick_main1.drop(columns = 9)
```

```python
blick_main1[9] = blick_main1[7]
blick_main1[10] = blick_main1[8]
```

```python
#replace topics with ressort-categories
with open("ressorts_cat.bin", "rb") as data:
    ressorts = pickle.load(data)

blick_main1[9].replace(ressorts, inplace=True)
blick_main1[10].replace(ressorts, inplace=True)
```

```python
# removing values
blick_main1 = blick_main1.dropna(subset=[0])
blick_main1 = blick_main1.dropna(subset=[5])
blick_main1 = blick_main1.dropna(subset=[6])
```

```python
#delete rows where category is unworkable
with open("to_delete.bin", "rb")as data:
    to_delete = pickle.load(data)

blick_main1 = blick_main1[~blick_main1[7].isin(to_delete)]
blick_main1 = blick_main1[~blick_main1[8].isin(to_delete)]
```

```python
blick_main1[11] = np.where(blick_main1[9] == blick_main1[10], "", blick_main1[10])
blick_main1 = blick_main1.drop(columns = 10)
```

```python
blick_main1[12] = "blick"
```

```python
#replacing "Agentur" with either proper agency name or then Kürzel
tester2 = blick_main1.copy()
tester2[13] = blick_main1[5].str[-5:]
tester2[14] = tester2[13].apply(lambda x: x[1:4] if x[0] == "(" else "")

blick_main1[4] = np.where(tester2[14] != "", tester2[14], blick_main1[4])
```

```python
blick_main1[4] = blick_main1[4].str.replace("http\S+", "",regex = True)
blick_main1[4] = blick_main1[4].str.replace("Bild", "")
```

```python
blick_main1.to_csv("blick_artikel.csv", header = None, sep = "|", index = False)
```
