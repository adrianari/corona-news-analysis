```python
import pandas as pd
import numpy as np
```

```python
df_main = pd.read_csv("woz_art1.csv", delimiter = "|", header = None)
```

```python
#drop duplicates
df_main = df_main.drop_duplicates(subset = [0], keep = "last")
```

```python
#remove columns without title, text or url
df_main = df_main.dropna(subset=[0])
df_main = df_main.dropna(subset=[5])
df_main = df_main.dropna(subset=[4])
```

```python
#extracting categories
categories = df_main[5].str.split("/",expand=True)
len(categories[4].unique())  ## we cannot extract much category information for WOZ in this regard
```

```python
#setting the date
df_main[2] = pd.to_datetime(df_main[2])
df_main[6] = df_main[2].dt.year
df_main = df_main[df_main[6] != 2021]
```

```python
#reset index
df_main.reset_index(drop=True)
```

```python
df_main[7] = ""
df_main[8] = ""
df_main[9] = ""
df_main[10] = ""
df_main[11] = ""
```

```python
df_main = df_main[[0,1,2,7,3,4,5,8,9,10,11]]
```

```python
df_main[12] = "woz"
df_main[4] = df_main[4].str.replace("http\S+", "",regex = True)
df_main[4] = df_main[4].str.replace("Bild", "")
```

```python
df_main.to_csv("woz_artikel.csv", header = None, sep = "|", index = False)
```
