```python
import pandas as pd
import numpy as np
import pickle
import re
```

```python
blick1 = pd.read_csv("watson_kom1.csv", delimiter = "{", header = None)
blick2 = pd.read_csv("watson_kom2.csv", delimiter = "|", header = None)
blick3 = pd.read_csv("watson_kom3.csv", delimiter = "|", header = None)
blick4 = pd.read_csv("watson_kom4.csv", delimiter = "|", header = None)
blick5 = pd.read_csv("watson_kom5.csv", delimiter = "|", header = None)
blick6 = pd.read_csv("watson_kom6.csv", delimiter = "|", header = None)

frames = [blick1, blick2, blick3, blick4, blick5, blick6]

df_kom = pd.concat(frames, ignore_index= True)
```

```python
df_kom
```

```python
df_kom = df_kom.drop_duplicates(keep = "last")
df_kom = df_kom.reindex(columns = [0,2,1])
```

```python
cleaner = df_kom[1].str.split(r'\s*[a-z]', expand=True, n=1)
cleaner = cleaner[0].str.split(" ", expand = True)
```

```python
cleaner
```

```python
df_kom[1] = pd.to_datetime(cleaner[0], format="%d.%m.%Y")
df_kom[5] = df_kom[1].dt.year
df_kom = df_kom[df_kom[5] != 2021]
df_kom = df_kom.drop(columns = 5)
df_kom[4] = ""
```

```python
df_kom
```

```python
emoji_pattern = re.compile("["
        u"\U0001F600-\U0001F64F"  # emoticons
        u"\U0001F900-\U0001F980"
        u"\U0001F923"
        u"\U0001F92A"
        u"\U0001F917"
        u"\U0001F912"
        u"\U0001F976"
        u"\U0001F929"
        u"\U0001F300-\U0001F5FF"  # symbols & pictographs
        u"\U0001F680-\U0001F6FF"  # transport & map symbols
        u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                           "]+", flags=re.UNICODE)


df_kom[2] = df_kom[2].str.replace(emoji_pattern, "")
df_kom[2] = df_kom[2].str.replace("\n", " ")
df_kom[2] = df_kom[2].str.replace("\ÃŸ", "ss")

df_kom = df_kom.dropna(subset = [2])
df_kom = df_kom.dropna(subset = [0])
```

```python
df_kom[0] = df_kom[0].str.strip()
df_kom[2] = df_kom[2].str.strip()
df_kom[10] = "watson"
```

```python
#weekdays
weekdays = []
for index, rows in df_kom.iterrows():
    weekdays.append(rows[1].weekday())
    
df_kom[11] = weekdays
```

```python
#time category

df_kom[12] = ""
```

```python
df_kom[13] = df_kom[2].str.len()
```

```python
df_kom.to_csv("watson_kom.csv", header = None, sep = "|", index = False)
```
