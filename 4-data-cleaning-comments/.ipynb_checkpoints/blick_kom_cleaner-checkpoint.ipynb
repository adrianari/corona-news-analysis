```python
import pandas as pd
import numpy as np
import pickle
import re
```

```python
blick1 = pd.read_csv("blick_kom1.csv", delimiter = "|", header = None)
blick2 = pd.read_csv("blick_kom2.csv", delimiter = "|", header = None)
blick3 = pd.read_csv("blick_kom3.csv", delimiter = "|", header = None)
blick4 = pd.read_csv("blick_kom4.csv", delimiter = "|", header = None)
blick5 = pd.read_csv("blick_kom5.csv", delimiter = "|", header = None)
blick6 = pd.read_csv("blick_kom6.csv", delimiter = "|", header = None)
blick7 = pd.read_csv("blick_kom7.csv", delimiter = "|", header = None)
blick8 = pd.read_csv("blick_kom8.csv", delimiter = "|", header = None)
blick9 = pd.read_csv("blick_kom9.csv", delimiter = "|", header = None)
blick1v = pd.read_csv("blick_kom1_vm.csv", delimiter = "|", header = None)
blick2v = pd.read_csv("blick_kom2_vm.csv", delimiter = "|", header = None)
blick3v = pd.read_csv("blick_kom3_vm.csv", delimiter = "|", header = None)
blick4v = pd.read_csv("blick_kom4_vm.csv", delimiter = "|", header = None)
blick5v = pd.read_csv("blick_kom5_vm.csv", delimiter = "|", header = None)
blick6v = pd.read_csv("blick_kom6_vm.csv", delimiter = "|", header = None)
blick7v = pd.read_csv("blick_kom7_vm.csv", delimiter = "|", header = None)
blick8v = pd.read_csv("blick_kom8_vm.csv", delimiter = "|", header = None)

frames = [blick1, blick2, blick3, blick4, blick5, blick6, blick7, blick8, blick9,
         blick1v, blick2v, blick3v, blick4v, blick5v, blick6v, blick7v, blick8v]

df_kom = pd.concat(frames, ignore_index= True)
```

```python
df_kom
```

```python
df_kom = df_kom.drop_duplicates(keep = "last")
```

```python
date_indy = df_kom[1].str.split(",", expand = True)
```

```python
df_kom[3] = date_indy[0]
df_kom[4] = date_indy[1]
df_kom[3] = pd.to_datetime(df_kom[3], format="%d.%m.%Y")
df_kom[5] = df_kom[3].dt.year
df_kom = df_kom[df_kom[5] != 2021]
```

```python
df_kom = df_kom.drop(columns = [1,5])
```

```python
df_kom[4] = df_kom[4].str.replace(" Uhr", "")
```

```python
df_kom[5] = "blick"
df_kom
```

```python
#weekdays
weekdays = []
for index, rows in df_kom.iterrows():
    weekdays.append(rows[3].weekday())
```

```python
df_kom[7] = weekdays
```

```python
df_kom[2] = df_kom[2].str.strip()
```

```python
#remove emojis and other things

emoji_pattern = re.compile("["
        u"\U0001F600-\U0001F64F"  # emoticons
        u"\U0001F900-\U0001F980"
        u"\U0001F923"
        u"\U0001F92A"
        u"\U0001F917"
        u"\U0001F912"
        u"\U0001F976"
        u"\U0001F929"
        u"\U0001F300-\U0001F5FF"  # symbols & pictographs
        u"\U0001F680-\U0001F6FF"  # transport & map symbols
        u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                           "]+", flags=re.UNICODE)
df_kom[2] = df_kom[2].str.replace(emoji_pattern, "")
df_kom[2] = df_kom[2].str.replace("\n", " ")
df_kom[2] = df_kom[2].str.replace("\ß", "ss")
```

```python
df_kom = df_kom.dropna(subset = [2])
df_kom = df_kom.dropna(subset = [0])
```

```python
df_kom[8] = df_kom[4].str.split(":").str[0]

df_kom[10] = df_kom[2].str.len()
```

```python
df_kom.to_csv("blick_kom.csv", header = None, sep = "|", index = False)
```
